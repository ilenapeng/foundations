```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.head()
```

```python
# Where are people getting shot
df[df.gun_related == True].primary_cause.value_counts()
```

```python
# give me something where primary cause contains gunshot wound
df[df.primary_cause.str.contains("GUNSHOT WOUND", na=False)] \
    .primary_cause.value_counts()
```

```python
# Get the primary cause 
# get rid of "THE"
# and pull out every time it says "GUNSHOT WOUND OF"

df.primary_cause.str.replace('THE ','') \
    .str.extract("GUNSHOT WOUND OF (.*)") \
    .value_counts() \
    .to_frame('counts') \
    .reset_index() \
    .rename(columns={0: 'body_part'})

# columns = is the same thing as axis. so rename_axis also works
# df.primary_cause.str.extract("GUNSHOT WOUND OF (.*)").value_counts()
```

```python
# Change in homicides?
df.manner_of_death.value_counts()
```

```python
homicides_df = df[df.manner_of_death == 'HOMICIDE'].copy()
homicides_df.head()
```

```python
# df.date_of_incident.dt.year won't work bc this is a date time

# 04/17/2015 10:25:00 PM
# Matching 00/00/0000

# \d = digit -> # character -> is the same thing as [0-9]
# () is a capture group

homicides_df['year'] = df.date_of_incident.str.extract('\d\d/\d\d/(\d\d\d\d)', expand=False)
# this looks for {\d\d (two numbers)} / {\d\d (another two numbers)} / {4 numbers}
# expand=False stops it from returning a dataframe
```

```python
homicides_df.year.value_counts().sort_index().plot()
```

```python
# to convert year to a date in homicides df
homicides_df['incident_datetime'] = pd.to_datetime(homicides_df.date_of_incident)
homicides_df.head()
```

```python
# to do the same with the first df
# coerce turns errors into NaNs
df['incident_datetime'] = pd.to_datetime(df.date_of_incident, errors='coerce')
df.head()
```

```python
# counting homicides by year again, but with our new date time variable
homicides_df.incident_datetime.dt.year.value_counts()
```

```python
# filter to only after 2014, since that's when the dataset is supposed to start
homicides_df = homicides_df[homicides_df.incident_datetime.dt.year > 2014]
homicides_df.head()
```

```python
# homicides by year
homicides_df.incident_datetime.dt.year.value_counts().sort_index().plot(kind='bar')
```

```python
# homicides by month
homicides_df.incident_datetime.dt.month.value_counts() \
    .sort_index() \
    .plot(ylim=[0, 1000])
```

```python
# homicides by month and year
# resample allows you to group based on months
homicides_df.resample('M', on='incident_datetime').size()
```

```python
homicides_df.resample('M', on='incident_datetime').size().plot()
```

```python
# yearly: A for annual
homicides_df.resample('A', on='incident_datetime').size().plot()
```

```python
# weekly: W,
homicides_df.resample('W', on='incident_datetime').size().plot()
```

```python
# daily: D
homicides_df.resample('D', on='incident_datetime').size().plot()
```

```python
# quarterly: Q
homicides_df.resample('Q', on='incident_datetime').size().plot()
```

```python
# every 2 weeks: 2W
homicides_df.resample('2W', on='incident_datetime').size().plot()
```

```python
# % change using .pct_change()
homicides_df.resample('M', on='incident_datetime').size().pct_change()
```
